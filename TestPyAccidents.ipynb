```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/accidents.csv")

df = df[["AvyYear","Killed"]]
#df.query('State == "UT" and \'TravelMode\=="Ski"')
#df.query(' PrimaryActivity == "Inbounds Rider"').count()#TravelMode == "Ski" and PrimaryActivity == "Inbounds Rider"').count()
df = df.groupby(["Killed","AvyYear"]).count().reset_index()
#df = df.query(['AvyYear'] == 2000)
#State == "UT" and  

result = df.query('AvyYear > 2000')

#df

#rename colums
result
```

```python
test = pd.read_csv("data/accidents.csv")

test = test[["AvyYear","Killed"]]
result = test.groupby("AvyYear").count().reset_index()

result = result.query('AvyYear > 2000')
result
```
